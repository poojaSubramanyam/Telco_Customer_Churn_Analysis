```python
# Changing the total chargees data type from object to numeric type

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

```python
# Converting the target column and mapping to binary 1, 0

df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 2})
```

```python
# Churn vs Payment Method
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
def tenure_group(x):
    if x <= 12:
        return '0-1 Year'
    elif x <= 36:
        return '1-3 Years'
    elif x <= 60:
        return '3-5 Years'
    else:
        return '5+ Years'

df['tenure_group'] = df['tenure'].apply(tenure_group)
```

```python
df.groupby('tenure_group')['Churn'].mean()
```
